```python
import pandas as pd
import numpy as np
```

```python
url = 'https://fbref.com/pt/comps/38/10073/2020-Serie-B-estatisticas#all_stats_squads_playing_time'
```

```python
list_df_tables = pd.read_html(url)
len(list_df_tables)
```

```python
df1 = list_df_tables[2]
df1
```

```python
tabela = pd.DataFrame()
mdf1={'nomes':['Equipe','NJogadores','MediaIdade','Posse','GolsPro','Assistencias',
              'PenalConvertidos', 'PenalBatidos', 'CAmarelo','CVermelho'],
      'posicao':[0,1,2,3,8,9,11,12,13,14]}
tabela[mdf1['nomes']] = df1.iloc[:,mdf1['posicao']]
tabela
```
